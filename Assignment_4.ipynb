```python
def find_max(a, b, c):
    return max(a, b, c)

print(find_max(10, 25, 15))
```

```python
def distinct_list(lst):
    return list(set(lst))

print(distinct_list([1, 2, 2, 3, 4, 4]))
```

```python
def multiply_list(lst):
    result = 1
    for i in lst:
        result *= i
    return result

print(multiply_list([2, 3, 4]))
```

```python
n = int(input("Enter a number to calculate its factorial: "))

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(n))
```

```python
s = input("Enter a string to reverse: ")
def reverse_string(s):
    return s[::-1]

print(reverse_string(s))
```

```python
def check_range(num, start, end):
    return start <= num <= end

print(check_range(5, 1, 10))
```

```python
def even_numbers(lst):
    for i in lst:
        if i % 2 == 0:
            print(i)

even_numbers([1, 2, 3, 4, 5, 6])
```

```python
n = int(input("Enter a no: "))
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(n))
```

```python
s = input("Enter string: ")
def count_case(s):
    upper = 0
    lower = 0
    
    for char in s:
        if char.isupper():
            upper += 1
        elif char.islower():
            lower += 1
            
    print("Uppercase:", upper)
    print("Lowercase:", lower)

print(count_case(s))
```
